# Grabbing all Jordan links from StockX

```python
url = 'https://stockx.com/retro-jordans'
```

```python
chromedriver = '/Applications/chromedriver'
os.environ['webdriver.chrome.driver'] = chromedriver
driver = webdriver.Chrome(chromedriver)
driver.get(url)
```

```python
#Need to click the "Load More" Button to grab all the shoes
for i in np.arange(26): #There are 25 "Load More Buttons" on the adidasPage
    driver.find_element_by_xpath('//button[@class="btn btn-default"]').click()
    time.sleep(3)
```

```python
#Grab all the links on the Jordan Sneaker page
jordan_shoes = []
for i in driver.find_elements_by_xpath('//a[@class="tile browse-tile"]'):
    url = i.get_attribute('href')
    jordan_shoes.append(url)
```

```python
print('There are a total of {} shoes from Jordan page from StockX '.format(len(jordan_shoes)))
```

```python
def SneakerInfoScapper():
    
    try:
        name = driver.find_element_by_xpath('//h1[@class="name"]').text
    except:
        name = 0
        print('no name')
        
    try:
        color = driver.find_element_by_xpath('//div[@class="detail"][2]/span').text
    except:
        color = 0
        print('no color')
    
    try:
        retail = driver.find_element_by_xpath('//div[@class="detail"][3]/span').text
    except:
        retail = 0
        print('no retail price')
        
    try:
        release_date = driver.find_element_by_xpath('//div[@class="detail"][4]/span').text
    except:
        release_date = 0
        print('no release_date')
        
    try:    
        last_sell = driver.find_element_by_xpath('//div[@class="sale-value"]').text  
    except:
        last_sell = 0
        print('no last sell')
    
    try:    
        annual_high = driver.find_element_by_xpath('//div[@class="value-container"][1]/span[1]').text
    except:
        annual_high =0
        print('no annual high')
        
    try:
        annual_low = driver.find_element_by_xpath('//div[@class="value-container"][1]/span[2]').text
    except:
        annual_low = 0
        print('no annual high')
        
    try:
        volatility = driver.find_element_by_xpath('//div[@class="value-container"][1]/span[@class="value"]').text
    except:
        volatility =0
        print('no volatility')
        
    try:
        trade_range = driver.find_element_by_xpath('//div[@class="ds-range value-container"][1]/span').text
    except: 
        trade_range=0
        print('no trade range')
    
    try:
        price_info = driver.find_elements_by_xpath('//div[@class="gauges"]')
    except:
        price_info =0
        print('no price info')
    
    try:
        total_shoes_sold = driver.find_element_by_xpath('//div[@class="gauges"]//div[@class="gauge-container"][1]//div[@class="gauge-value"]').text
    except:
        total_shoes_sold =0
        print('no total shoes')
        
    try:
        price_premium = driver.find_element_by_xpath('//div[@class="gauges"]//div[@class="gauge-container"][2]//div[@class="gauge-value"]').text
    except:
        price_premium = 0
        print('no price prem')
        
    try:
        avg_sale_price = driver.find_element_by_xpath('//div[@class="gauges"]//div[@class="gauge-container"][3]//div[@class="gauge-value"]').text
    except:
        avg_sale_price = 0
        print('no avg price')
    
    
    
    return name,color,retail,release_date,last_sell,annual_high,annual_low,volatility,trade_range,total_shoes_sold,price_premium,avg_sale_price
       
```

```python
csv_file = open('Jordan_StockX.csv', 'w')
writer = csv.writer(csv_file)
writer.writerow(['Brand','Model','Color','Retail_Price', 'Release_Date', 'Last_Sell_Price', 'Annual_high','Annual_low','Volatility','Trade_Range','Yearly_shoes_sold','Price_Premium','Avg_Sale_Price'])
```

```python
chromedriver = '/Applications/chromedriver'
os.environ['webdriver.chrome.driver'] = chromedriver
driver = webdriver.Chrome(chromedriver)
```

```python
for each_url in jordan_shoes:
    driver.get(each_url)
    time.sleep(2)
    name,color,retail,release_date,last_sell,annual_high,annual_low,volatility,trade_range,total_shoes_sold,price_premium,avg_sale_price = SneakerInfoScapper()
    print(each_url)
    sneakers_dict ={}
    sneakers_dict['Brand'] = 'Nike'
    sneakers_dict['Model'] = name
    sneakers_dict['Color'] = color
    sneakers_dict['Retail_Price'] = retail
    sneakers_dict['Release_Date'] = release_date
    sneakers_dict['Last_Sell_Price'] = last_sell
    sneakers_dict['Annual_high'] = annual_high
    sneakers_dict['Annual_low'] = annual_low
    sneakers_dict['Volatility'] = volatility
    sneakers_dict['Trade_Range'] = trade_range
    sneakers_dict['Yearly_shoes_sold'] = total_shoes_sold
    sneakers_dict['Price_Premium'] = price_premium
    sneakers_dict['Avg_Sale_Price'] = avg_sale_price
    writer.writerow(sneakers_dict.values())


csv_file.close()
driver.close()
```
